```python
import os
import glob
from collections import Counter

# Load all csv files from the specified folder
folder_path = '../Datasets/Dataset_2/SensorData'
csv_files = glob.glob(os.path.join(folder_path, '**', '*.csv'), recursive=True)
print(len(csv_files))

# Extract first number from filename (before underscore)
numbers = []
for file in csv_files:
    filename = os.path.basename(file)
    first_number = filename.split('_')[0]
    numbers.append(first_number)

# Count occurrences of each number
count = Counter(numbers)

# Print numbers that don't appear exactly 4 times
anomalies = [num for num, cnt in count.items() if cnt != 4]
print("Numbers that do not appear 4 times:")
for num in anomalies:
    print(f"  {num}: {count[num]} times")
```
